```python
import os, sys

#To use Spark in Python rather in Scala  
from pyspark import SparkContext
from pyspark.sql import SparkSession
from copy import copy 
import time 

#Iterate mapper

def ccf_iterate_map(X):
    return [(X[0], X[1]), (X[1], X[0])]

#Iterate deducer 

def cff_iterate_reduce(X):
    
    #Global access to PAIR_COUNTER
    global PAIR_COUNTER
  
    key = X[0]
    values = X[1]
    L = []
    valueList = []
    minim = key

    for value in values:
        if value<minim:
            minim=value
            valueList.append(value)

    if minim<key:
        
        L.append((key, minim))
        for value in valueList:
            if minim != value:
                
                PAIR_COUNTER += 1
                L.append((value, minim))

    return L

#Auto-execution

if __name__ == "__main__":
    
  
    start_time = time.time()
    nbiter = 0
  
  #To use Spark
  
    sc = pyspark.SparkConf()
    PAIR_COUNTER = sc.accumulator(1)
   
  # Input file as "start_node end_node"
  
    lines = sc.textFile("Web-test.txt")
    rows = lines.map(lambda x: x.strip().split())
    graphNodes = rows.map(lambda x: (int(x[0]), int(x[1])))
    print(graphNodes)
    OLD_COUNTER = -1
  
  
  #Stop condition of the loop 
  
    while OLD_COUNTER != PAIR_COUNTER.value:
        nbiter += 1
        OLD_COUNTER = PAIR_COUNTER.value
    
    #Iterate Map Reduce
    
    graphNodes = graphNodes.flatMap(ccf_iterate_map)
    graphNodes = graphNodes.map(lambda x : (x[0], [x[1]])).reduceByKey(lambda a,b : a + b)
    graphNodes = graphNodes.flatMap(cff_iterate_reduce)
    # Dedup MapReduce
    graphNodes = graphNodes.distinct()
    graphNodes = graphNodes.sortBy(lambda x: x[1], True)
    graphNodes = graphNodes.sortBy(lambda x: x[0], True)

    print("GraphNodes")
    print(graphNodes)
    
    # to create the output file
    L = graphNodes.collect()
    print(L)
    print("DARIO")
    f = open("outputwG.txt", "w") 
    for elem in L:
        f.write("{} {}\n".format(elem[0], elem[1])) 
        print("OK")
        print(elem[0], elem[1])
    f.close()
    # to close the Spark environment
    sc.stop()
    print("Execution time: %s s" % (time.time() - start_time)) 
    print("Number of iterations: %s" %(nbiter))
    
```
